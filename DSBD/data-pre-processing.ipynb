```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
```

```python
file_path = 'titanic.csv'
try:
    df = pd.read_csv(file_path)
    print(f"\nDataset '{file_path}' loaded sucessfully intp a dataframe")

except FileNotFoundError:
    print(f"\nError: File '{file_path}' not found. Please check the file path.")
```

```python
#handling missing values for 'Age'
missing_values = df.isnull().sum()
print("Missing Values: \n", missing_values[missing_values > 0])
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace = True)
print(f"Age coloumn inputed with median: {median_age}")
```

```python
#embarked

mode_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_embarked, inplace = True)
print(f"Embarked coloumn inputed with mode: {mode_embarked}")
```

```python
#cabin

df.drop('Cabin', axis = 1, inplace = True)
print("Cabin coloumn dropped")
```

```python
print("Missing values after cleaning: \n", df.isnull().sum()[df.isnull().sum() > 0])
```

```python
#Outliers 

Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 - 1.5*IQR
```

```python
df['Fare'] = np.where(df['Fare'] > upper_bound, upper_bound, df['Fare'])
df['Fare'] = np.where(df['Fare'] < lower_bound, lower_bound, df['Fare'])
```

```python
# data-transformation

df['Fare_Log'] = np.log1p(df['Fare'])
```

```python
# cateogarical variables

categorical_cols = ['Sex', 'Embarked']

# Use One-Hot Encoding for nominal variables
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
print(f"One-Hot Encoding applied to {categorical_cols} and 'Pclass'.")
print("First few rows of the final transformed DataFrame:")
print(df[['Sex_male', 'Embarked_Q', 'Embarked_S', 'Fare', 'Fare_Log']].head())
```
